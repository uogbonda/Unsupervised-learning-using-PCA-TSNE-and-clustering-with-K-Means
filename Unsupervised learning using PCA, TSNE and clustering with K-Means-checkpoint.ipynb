```python
def pca(X=np.array([]), ndim=2):
    print("Standardizing the data first and then reducing the dimension with PCA ...")
    (n, d) = X.shape
    X_sc = StandardScaler().fit_transform(X)
    X = X_sc - np.tile(np.mean(X_sc, 0), (n, 1))
    (A, B) = np.linalg.eig(np.dot(X.T, X))
    Y = np.dot(X, B[:, 0:ndim])
    return Y
```

```python
def entropy(D, inv_sigma=1.0):
    eps=1e-7
    P = np.exp(-D * inv_sigma) 
    sumP = np.sum(P)
    P=np.divide(P,sumP, where=sumP!=0)
    a=0.2
    E = (-1/(a-1))* np.log(np.sum(P**a) + eps)
    #P = P / sumP
    return E, P
```

```python
def search_data_perp(X, tolerance=1e-7, perp=20.0):
    (n, d) = X.shape
    print(f'Computing the pairwise distances of the data points: {n}')
    # perp is perplexity
    #(n, d) = X.shape
    sumX = np.sum(np.square(X), 1)
    s=np.add(-2 * np.dot(X, X.T),sumX).T
    D = np.add(s, sumX)
    P = np.zeros((n, n))
    inv_sigma = np.ones((n, 1))
    logp = np.log(perp)

    # Performing a loop where with all the points of the data
    for i in range(n):
        if i % 500 == 0:
            print(f' Computing P-values for point {i} of {n}')
            if i % n == 0:
                print(f' Computing total P-values for all points {(i,n)}')  
        # Computation that handles entropy and precision
        min_invsig = -np.inf
        max_invsig = np.inf
        Di = D[i, np.concatenate((np.r_[0:i], np.r_[i+1:n]))]     # np.r_ changes the sliced points to the concatenate in numpy library
        (E, thisP) = entropy(Di, inv_sigma[i])

        # Evaluate whether the perplexity is within tolerance given here
        E_diff = E - logp
        tries = 0
        while np.abs(E_diff) > tolerance and tries < 50:
            if E_diff > 0:
                min_invsig = inv_sigma[i].copy()
                if max_invsig == np.inf or max_invsig == -np.inf:
                    inv_sigma[i] = inv_sigma[i] * 2.
                else:
                    inv_sigma[i] = (inv_sigma[i] + max_invsig) / 2.
            else:
                max_invsig = inv_sigma[i].copy()
                if min_invsig == np.inf or min_invsig == -np.inf:
                    inv_sigma[i] = inv_sigma[i] / 2.
                else:
                    inv_sigma[i] = (inv_sigma[i] + min_invsig) / 2.

            # Recompute the values
            (E, thisP) = entropy(Di, inv_sigma[i])
            E_diff = E - logp
            tries += 1

        # Set the final row of P
        P[i, np.concatenate((np.r_[0:i], np.r_[i+1:n]))] = thisP

    # Return final P-matrix
    print(f' The mean value of sigma is {round(np.mean(np.sqrt(1 / inv_sigma)),6)}')
    return P

```

```python
def tsne(X, ndim=2, init_dim=4, perp=20.0, lr=150, n_iter=1000):
    
    # to know if the inputs are ok.
    if isinstance(ndim, float):
        print(f'Error has occurred: array X should have type float.')
        return -1
    if round(ndim) != ndim:
        print(f'Error has occurred: the number of dimensions should be an integer.')
        return -1

    # Initialize variables
    X = pca(X, init_dim)  
    (n, d) = X.shape
    a=0.2
    initial_momentum = 0.5
    final_momentum = 0.8
                                          
    min_gain = 0.01
    Y = np.random.randn(n, ndim)

    dY = np.zeros((n, ndim))
    Y_i = np.zeros((n, ndim))
    gains = np.ones((n, ndim))

    # Computing the values of P
    P = search_data_perp(X, 1e-5, perp)
    P = P + np.transpose(P)
    P = P / np.sum(P)
    P = P * 4.									# early exaggeration initially 4
    P = np.maximum(P, 1e-12)

    # Run iterations
    for j in range(n_iter):

        # Evaluating the pairwise calculation and , we get Student t-distribution
        # Compute pairwise affinities
        sumY = np.sum(np.square(Y), 1)
        num = -2. * np.dot(Y, Y.T)
        num = 1. / (1. + np.add(np.add(num, sumY).T, sumY))
        num[range(n), range(n)] = 0.
        Q = num / np.sum(num)
        Q = np.maximum(Q, 1e-12)

        # Compute gradient
        PQ=((P**a)*(Q**(1-a)))/np.sum((P**a)*(Q**(1-a)))
        PQ=PQ-Q
        for i in range(n):
            dY[i, :] = np.sum(np.tile(PQ[:, i] * num[:, i], (ndim, 1)).T * (Y[i, :] - Y), 0)

        # Below is the update.....'j' is counting number of 1 to the maximum iteration
        #j=0
        j=int(j)
        if j < 20:
            momentum = initial_momentum
        else:
            momentum = final_momentum
        gains = (gains + 0.2) * ((dY > 0.) != (Y_i > 0.)) + (gains * 0.8) * ((dY > 0.) == (Y_i > 0.))
        gains[gains < min_gain] = min_gain
        Y_i = momentum * Y_i - lr * (gains * dY)
        Y = Y + Y_i
        Y = Y - np.tile(np.mean(Y, 0), (n, 1))

        # Compute current value of cost function
        if (j + 1) % 250 == 0:
            #R = np.sum(P * np.log(P / Q))                # if (iter +1 )% 10==0 before
            R = (1/(a-1))*np.log(np.sum((P**a) / (Q**(a-1))))  
            print(f'The iteration {j+1} and the error {round(R,6)}') 
            #print("Iteration %d: error is %f" % (iter + 1, R))

        # Remove the early exaggeration used previously which was 4.
        if j == 100:
            P = P / 4.

    # Return solution
    return Y
```
